```python
import requests
import pandas as pd
```

```python
response = requests.get("https://fakestoreapi.com/products")
```

```python
df = pd.DataFrame(response.json())[["title", "category", "price"]]
df.head()
```

```python
rate = []
```

```python
for i in range(0,len(response.json())):
  rate.append(response.json()[i]["rating"]["rate"])
print(len(rate))
```

```python
df["rate"] = rate
```
